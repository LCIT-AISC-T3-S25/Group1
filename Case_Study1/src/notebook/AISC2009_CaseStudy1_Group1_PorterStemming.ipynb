## Imports

```python
import pandas as pd
import re
from nltk.tokenize import TweetTokenizer
from nltk.stem import PorterStemmer
```

## Reading Data

```python
df = pd.read_csv("Data.csv")
```

```python
df.head
```

## Concatenate all the text entries

```python
text_data = " ".join(df['Data'].dropna().astype(str))
```

## Tokenize data

```python
tokens = re.findall(r'\b[a-zA-Z]+\b', text_data.lower())
```

```python
unique_before = set(tokens)
```

## Perform Stemming

```python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in tokens]
```

```python
unique_after = set(stemmed_tokens)
```

```python
print("Unique tokens before stemming:", len(unique_before))
print("Unique tokens after stemming:", len(unique_after))
```

```python
print("First 20 stemmed tokens:")
print(stemmed_tokens[:20])
```

### While doing this method, the stemmer is skipping numbers. So, to treat them. 

```python
tokens = re.findall(r"\b\w[\w'.:/@-]*\b", text_data)
unique_before = set(tokens)
```

```python
stemmed_tokens = [
    stemmer.stem(token) if token.isalpha() else token
    for token in tokens
]

unique_after = set(stemmed_tokens)
```

```python
print("Unique tokens before stemming:", len(unique_before))
print("Unique tokens after stemming:", len(unique_after))
print("First 20 stemmed tokens:")
print(stemmed_tokens[:20])
```
