**Factors to consider**
* Multiple newlines \n\n\n	May create tokenization artifacts or unintended paragraph breaks.
* Trailing/leading newlines	Can cause padding issues in prompt formatting.
* HTML <br> used instead	Missed by simple \n regex — should optionally detect.
* Invisibles like \r\n (Windows)	Could lead to false negatives if not normalized.
* Text split mid-sentence	Can affect context chunking or coherence in LLM summarization.


```
# Define a pattern to match all newline variants
newline_pattern = r'(?:\r\n|\r|\n|<br\s*/?>)'

# Count number of newlines in each record
df_no_duplicates['newline_count'] = df_no_duplicates['Data'].str.count(newline_pattern, flags=re.IGNORECASE)

# Count total records with at least one newline
has_newline_mask = df['newline_count'] >= 1
total_with_newlines = has_newline_mask.sum()

# Count total records with 2 or more newlines
multi_newline_mask = df_no_duplicates['newline_count'] >= 2
multi_newline_count = multi_newline_mask.sum()

# Print results
print(f"✅ Total records with at least one newline: {total_with_newlines}")
print(f"✅ Records with 2 or more newlines: {multi_newline_count}")

# Optional: preview those records
print("\n🔍 Sample of records with 2+ newlines:")
print(df_no_duplicates[multi_newline_mask][['Data', 'newline_count']].head())
```

## Q.No.  A.6

#### What is the total number of ip addresses across all the records?

```
text_data = ' '.join(language_data['Data'].dropna().astype(str))

# Regex patterns for different IP formats
ipv4_pattern = r'\b(?:(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\.){3}(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\b'
cidr_pattern = r'\b(?:(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\.){3}(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)/[0-9]{1,2}\b'
ipv6_full = r'\b(?:[A-Fa-f0-9]{1,4}:){7}[A-Fa-f0-9]{1,4}\b'
ipv6_compressed = r'\b(?:[A-Fa-f0-9]{1,4}:){1,7}:|::(?:[A-Fa-f0-9]{1,4}:){1,7}[A-Fa-f0-9]{1,4}\b'
ipv6_mapped_ipv4 = r'\b::ffff:(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\.(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\.(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\.(?:25[0-5]|2[0-4]\d|1\d{2}|[1-9]?\d)\b'

# Additional non-standard IPv4 formats
ipv4_integer = r'\b\d{8,10}\b'                      # 32-bit decimal integer
ipv4_hex = r'\b0x[0-9A-Fa-f]{8}\b'                  # Hexadecimal
ipv4_octal = r'\b0[0-7]+\.[0-7]+\.[0-7]+\.[0-7]+\b' # Octal

# Combine all regexes
all_patterns = f"{ipv4_pattern}|{cidr_pattern}|{ipv6_full}|{ipv6_compressed}|{ipv6_mapped_ipv4}|{ipv4_integer}|{ipv4_hex}|{ipv4_octal}"
matches = re.findall(all_patterns, text_data)

# Normalize and validate
valid_ips = []

for match in matches:
    ip = next(filter(None, match))  # Get non-empty group from match tuple
    try:
        # Handle hex (e.g., 0xC0A80101)
        if ip.startswith("0x"):
            ip = ipaddress.IPv4Address(int(ip, 16)).compressed
        # Handle decimal integer
        elif re.fullmatch(ipv4_integer, ip):
            ip = ipaddress.IPv4Address(int(ip)).compressed
        # Handle octal
        elif re.fullmatch(ipv4_octal, ip):
            parts = [str(int(part, 8)) for part in ip.split('.')]
            ip = '.'.join(parts)
            ipaddress.IPv4Address(ip)  # validate
        else:
            ipaddress.ip_network(ip, strict=False)
        valid_ips.append(ip)
    except ValueError:
        continue


print(f"Total valid IP addresses (all formats): {len(valid_ips)}")
```

#### Q2. How many records have a word starting with the letter “w”?

#### Q3. How many records make a word that starts with an alphabet and is not a URL?

```
# Defining a pattern to find all words starting with an alphabet
word_pattern = re.compile(r'\b[a-zA-Z][\w-]*\b')

# List of common domain extensions to filter
domain_extensions = ['com', 'org', 'net', 'io', 'gov', 'edu', 'ca', 'co', 'us', 'uk', 'info', 'biz', 'me']

# Pattern to detect domains
domain_regex = re.compile(r'\b[\w.-]+\.(?:' + '|'.join(domain_extensions) + r')\b', re.IGNORECASE)
```

```
# Function to check if record contains a valid word that is not a domain or URL
def has_valid_non_url_word(text):
    if not isinstance(text, str):
        return False
    if domain_regex.search(text):
        return False
    words = word_pattern.findall(text)
    return len(words) > 0

# Filter records
valid_records = df_string[df_string.apply(has_valid_non_url_word)]
valid_records.head()
```

```
# Print number of valid records
print("Count of records that make a word that starts with an alphabet and is not a URL : ", len(valid_records))
```

#### Result & Observation(s)

### 1787 Records
The regular expression strictly filters all urls from the given data, including most common domains.
