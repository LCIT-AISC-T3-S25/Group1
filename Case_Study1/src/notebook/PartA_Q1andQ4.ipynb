Q.No.1 How many records have a date that is expressed without using
alphabets?

```python
## checking for dates in numeric formats
date_pattern = r'\b(?:\d{1,4}[-/.]){2}\d{1,4}\b'
# Check each row for a date using regex (returns True/False)
data_v1['Has_Date'] = data_v1['Data'].str.contains(date_pattern)
data_v1['Extracted_Date'] = data_v1['Data'].str.extract(f'({date_pattern})')

# Show result
print(data_v1)

```

```python

##checking whether the row has dates 
# Filter rows where a date was detected (Has_Date == True)
true_cases = data_v1[data_v1['Has_Date'] == True]
# Display number of true cases and sample results
print("Total rows with numeric dates:", len(true_cases))
print(true_cases.head(16))
```

These are phone or toll free numbers not real dates.

```python
## removing phone or toll free numbers
phone_pattern = r'^\d{3}[-.]\d{3}[-.]\d{4}$'
toll_free_pattern = r'^1[-.]\d{3}[-.]\d{3}$'
emergency_pattern = r'^\d{1,2}[-.]\d{1,2}[-.]\d{1,2}$'  

# Remove matches that are likely false positives
df_cleaned = data_v1[
    ~data_v1['Extracted_Date'].str.match(phone_pattern, na=False) &
    ~data_v1['Extracted_Date'].str.match(toll_free_pattern, na=False) &
    ~data_v1['Extracted_Date'].str.match(emergency_pattern, na=False)
]

# Step 5: Filter for valid dates only
valid_dates = df_cleaned[df_cleaned['Has_Date'] == True]

# View result
print("Number of dates found:", len(valid_dates))
print(valid_dates[['Data', 'Extracted_Date']].head())
```

Q.No.4 How many tweets contain one of these emojis :), :D, ;), :P?

```python
# Define regex pattern for the emojis
emoji_pattern = r'(:\)|:D|;\)|:P)'
# Check if each row contains any of the emojis
data_v1['Has_Emoji'] = data_v1['Data'].astype(str).str.contains(emoji_pattern)
print(data_v1[data_v1['Has_Emoji']].head())
# Count the number of matching tweets
emoji_count = data_v1['Has_Emoji'].sum()
print("Number of tweets containing emojis :), :D, ;), or :P =", emoji_count)

```

```python
##Cross checking the results using re module
# Define the regex pattern for the emojis
emoji_pattern = re.compile(r'(:\)|:D|;\)|:P)')

# Define a function to check for emoji presence in a string
def contains_emoji(text):
    return bool(emoji_pattern.search(str(text)))  # convert to string and search

# Apply the function to each row in the Data column
data_v1['Has_Emoji_1'] = data_v1['Data'].apply(contains_emoji)

# Count and show results
emoji_count = data_v1['Has_Emoji_1'].sum()
print("Number of tweets with emojis :), :D, ;), or :P =", emoji_count)


```
