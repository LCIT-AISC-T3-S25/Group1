#### Q2. How many records have a word starting with the letter “w”?

```python
# Defining a regular expression to include words starting with "w" (case sensitive)
match_pattern_w = r'\bw[a-zA-Z]*\b'
```

```python
# Filter matching rows
matching_rows = df_string[df_string.str.contains(match_pattern_w, regex=True)]

# A sample output of 5 matched rows 
sample_matches = matching_rows.head(5).tolist()

for i, row in enumerate(sample_matches, 1):
    print(f"{i}: {row}")
```

```python
# Count of records matching the defined pattern
count_w = df_string.str.contains(match_pattern_w, regex=True).sum()
print("Records with a word starting with 'w':", count_w)
```
