```python
import pandas as pd;
import re;
```

```python
df = pd.read_csv("Data.csv")
```

```python
df.head(5)
```

```python
count_before = 0
count_after = 0
```

```python
for i in df['Data']:
    text = str(i)
    # Count of non-alphanumerics before substitution
    count_before += len(re.findall(r'[^a-zA-Z0-9]', text))
```

```python
converted_data =[]

for i in df['Data']:
    text = str(i)
    
    converted_values = re.sub(r'[^a-zA-Z0-9]', '\n', text)
    converted_data.append(converted_values)
    
    # Count of non-alphanumerics after substitution
    count_after += len(re.findall(r'[^a-zA-Z0-9\n]', converted_values))
```

```python
df['converted'] = converted_data
```

```python
print(df[['Data', 'converted']].head(5))
```

```python
# Print total counts
print("Total non-alphanumeric characters BEFORE substitution:", count_before)
print("Total non-alphanumeric characters AFTER substitution(excluding newline):", count_after)
```
