```python
import pandas as pd
import re
```

```python
# Load your CSV file
df = pd.read_csv('Data.csv')
```

```python
# Define URL matching regex pattern
url_pattern = r'https?://\S+|www\.\S+|\b\w+\.\w{2,}(?:/\S*)?'
```

```python
# Function to count URLs in a single text entry
def count_urls(text):
    if isinstance(text, str):
        return len(re.findall(url_pattern, text))
    return 0
```

## Total Number of URLs count

```python
# Apply the function to every cell and sum the total
total_urls = df.map(count_urls).sum().sum()

print("Total number of URLs in the data:", total_urls)
```

## Counting total number of distinct URLs

```python
# Function to extract URLs from a single text entry
def extract_urls(text):
    if isinstance(text, str):
        return re.findall(url_pattern, text)
    return []

# Extract all URLs from every cell in the DataFrame
all_urls = df.map(extract_urls).values.flatten()

# Flatten the list of URL lists and filter out non-lists
flat_url_list = [url for sublist in all_urls if isinstance(sublist, list) for url in sublist]
```

## Total Number of Distinct URLs

```python
# Convert to set to get unique URLs
unique_urls = set(flat_url_list)

# Print the number of unique URLs
print("Total Number of unique (distinct) URLs:", len(unique_urls))
```
