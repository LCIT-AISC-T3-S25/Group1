```python
import nltk
import re
import string
from collections import Counter
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Download only necessary resources
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')  # Open Multilingual WordNet

def preprocess_text(text):
    """
    Preprocess text by converting to lowercase, removing URLs,
    punctuation, special characters, and extra whitespace
    """
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text

def custom_tokenize(text):
    """
    Tokenize text using a simple approach without NLTK's word_tokenize
    """
    # Split by whitespace
    words = text.split()

    # Filter out short tokens
    words = [word for word in words if len(word) > 1]

    return words

def remove_stopwords(tokens):
    """
    Remove stopwords from tokens, Bcause they occur more frequently and provide little semantic value
    """
    try:
        from nltk.corpus import stopwords
        stop_words = set(stopwords.words('english'))
    except:
        print('Manually removing stopwrds......')
        # Fallback to a basic stopwords list If NLTK is not working
        stop_words = {'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you',
                     'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his',
                     'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself',
                     'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which',
                     'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are',
                     'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having',
                     'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if',
                     'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for',
                     'with', 'about', 'against', 'between', 'into', 'through', 'during',
                     'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in',
                     'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once',
                     'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both',
                     'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor',
                     'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very'}

    return [word for word in tokens if word.lower() not in stop_words]

def tokenize_and_clean(text):
    """
    Preprocess, tokenize, and remove stopwords
    """
    # Preprocess the text
    processed_text = preprocess_text(text)

    # Tokenize the text
    tokens = custom_tokenize(processed_text)

    # Remove stopwords
    tokens = remove_stopwords(tokens)

    return tokens

def apply_stemming(tokens):
    """
    Apply Porter stemming to tokens
    """
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

def apply_lemmatization(tokens):
    """
    Apply NLTK lemmatization to tokens
    """
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens

def main():
    # Read the CSV file
    try:
        # Read the file as text
        with open('Data.csv', 'r', encoding='utf-8') as file:
            data = file.read()
    except UnicodeDecodeError:
        # Try with a different encoding if utf-8 fails
        with open('Data.csv', 'r', encoding='latin-1') as file:
            data = file.read()

    # Tokenize and clean the text
    tokens = tokenize_and_clean(data)

    # Apply Porter stemming
    stemmed_tokens = apply_stemming(tokens)

    # Apply NLTK lemmatization
    lemmatized_tokens = apply_lemmatization(tokens)

    # Count unique tokens before and after stemming/lemmatization
    original_unique = len(set(tokens))
    stemmed_unique = len(set(stemmed_tokens))
    lemmatized_unique = len(set(lemmatized_tokens))

    # Print results for stemming
    print("\nB.1 - Porter Stemming Results:")
    print(f"Original tokens count: {len(tokens)}")
    print(f"Original unique tokens count: {original_unique}")
    print(f"Stemmed tokens count: {len(stemmed_tokens)}")
    print(f"Stemmed unique tokens count: {stemmed_unique}")
    print(f"Reduction in unique tokens after stemming: {original_unique - stemmed_unique} ({(100 * (original_unique - stemmed_unique) / original_unique):.2f}%)")

    # Print results for lemmatization
    print("\nB.2 - NLTK Lemmatization Results:")
    print(f"Original tokens count: {len(tokens)}")
    print(f"Original unique tokens count: {original_unique}")
    print(f"Lemmatized tokens count: {len(lemmatized_tokens)}")
    print(f"Lemmatized unique tokens count: {lemmatized_unique}")
    print(f"Reduction in unique tokens after lemmatization: {original_unique - lemmatized_unique} ({(100 * (original_unique - lemmatized_unique) / original_unique):.2f}%)")

    # B.3 - Compare word frequencies
    original_freq = Counter(tokens)
    stemmed_freq = Counter(stemmed_tokens)
    lemmatized_freq = Counter(lemmatized_tokens)

    print("\nB.3 - Word Frequency Comparison:")

    print("\nTop 10 Original Words:")
    for word, count in original_freq.most_common(10):
        print(f"{word}: {count}")

    print("\nTop 10 Stemmed Words:")
    for word, count in stemmed_freq.most_common(10):
        print(f"{word}: {count}")

    print("\nTop 10 Lemmatized Words:")
    for word, count in lemmatized_freq.most_common(10):
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
```
