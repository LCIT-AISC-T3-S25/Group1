Question 5) How many records contain a decimal number?

```python
import pandas as pd
import re
# Load the uploaded CSV file
file_path = 'Downloads/Data.csv'
df = pd.read_csv(file_path)

# Display the first few rows and the column names to understand the structure
df.head(), df.columns
```

```python
# Regex to find decimal numbers
decimal_pattern = r'\b\d+\.\d+\b'

# Count records containing decimal numbers
decimal_count = df['Data'].str.contains(decimal_pattern, regex=True).sum()

print("Number of records with a decimal number:", decimal_count)
```

```python
# Filter rows with decimal numbers
decimal_rows = df[df['Data'].str.contains(r'\b\d+\.\d+\b', regex=True)]

# Display them
decimal_rows
```

```python
# Extract and display records with decimal matches
df['decimal_matches'] = df['Data'].apply(lambda x: re.findall(r'\b\d+\.\d+\b', str(x)))
decimal_rows = df[df['decimal_matches'].map(len) > 0]

decimal_rows[['Data', 'decimal_matches']]
```

**Therefore, 25 records contain decimals**
